Верхний график (price_per_meter vs floors_count)
Описание: На этом графике по оси X отложено количество этажей здания, а по оси Y — цена за квадратный метр.
Анализ:
Видно, что большинство объектов сосредоточено на меньшем количестве этажей, примерно до 20–30 этажей.
В зданиях с большим числом этажей (от 40 и выше) есть отдельные объекты с высокой стоимостью, но они представлены в меньшем количестве.
В целом, здесь нет четкой связи между числом этажей и ценой за квадратный метр: цена варьируется в довольно широких пределах при любом числе этажей.

Нижний график (price_per_meter vs rooms_count)
Описание: По оси X указано количество комнат в квартире, а по оси Y — цена за квадратный метр.
Анализ:
Видно, что цена за квадратный метр варьируется в каждой категории комнат, но каких-то ярко выраженных зависимостей между количеством комнат и стоимостью не наблюдается.
Каждое количество комнат имеет примерно одинаковый диапазон значений по цене за квадратный метр, что говорит о том, что количество комнат не является значимым фактором для изменения стоимости за квадратный метр.

```python
import numpy as np

# Удаление строк с отрицательными значениями количества комнат или этажей
df = df[(df['rooms_count'] >= 0) & (df['floors_count'] >= 0)]

# Рассчитываем IQR для 'price_per_meter'
Q1 = df['price_per_meter'].quantile(0.25)
Q3 = df['price_per_meter'].quantile(0.75)
IQR = Q3 - Q1

# Фильтруем выбросы по правилу 1.5 * IQR
df = df[(df['price_per_meter'] >= Q1 - 1.5 * IQR) & (df['price_per_meter'] <= Q3 + 1.5 * IQR)]

# Заменяем оставшиеся аномальные значения на NaN (например, цены, превышающие верхний порог)
df['price_per_meter'] = np.where(df['price_per_meter'] > 1e8, np.nan, df['price_per_meter'])

# Находим числовые столбцы
numeric_columns = df.select_dtypes(include=[np.number]).columns

# Заполняем NaN медианой в числовых столбцах
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].median())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 8))
sns.boxplot(data=df, x='district', y='price_per_meter')
plt.xticks(rotation=90)
plt.title('Распределение цены за квадратный метр по районам')
plt.xlabel('Район')
plt.ylabel('Цена за квадратный метр')
plt.show()
```

```python
# Получаем топ-10 районов по медианной цене за квадратный метр
top_districts = df.groupby('district')['price_per_meter'].median().nlargest(10).index
filtered_df = df[df['district'].isin(top_districts)]

plt.figure(figsize=(12, 6))
sns.boxplot(data=filtered_df, x='district', y='price_per_meter')
plt.xticks(rotation=90)
plt.title('Топ-10 районов по цене за квадратный метр')
plt.xlabel('Район')
plt.ylabel('Цена за квадратный метр')
plt.show()
```

```python
# Зависимость цены за квадратный метр от станции метро (underground)
plt.figure(figsize=(15, 8))
sns.boxplot(data=df, x='underground', y='price_per_meter')
plt.xticks(rotation=90)
plt.title('Распределение цены за квадратный метр по станциям метро')
plt.xlabel('Станция метро')
plt.ylabel('Цена за квадратный метр')
plt.show()
```

```python
# Получаем топ-10 станций метро по медианной цене за квадратный метр
top_underground = df.groupby('underground')['price_per_meter'].median().nlargest(10).index
filtered_df = df[df['underground'].isin(top_underground)]

plt.figure(figsize=(12, 6))
sns.boxplot(data=filtered_df, x='underground', y='price_per_meter')
plt.xticks(rotation=90)
plt.title('Топ-10 станций метро по цене за квадратный метр')
plt.xlabel('Станция метро')
plt.ylabel('Цена за квадратный метр')
plt.show()
```
